```python
#Files to iterate throught the CSV 

import numpy as np
import pandas as pd
import math
data = pd.read_csv("C:\\Users\\Ryan\\Downloads\\ResQv6.csv", error_bad_lines = False)
                  
data.head()
```

```python


dictionary = {"EquipmentName": data.EquipmentName, "ModelNumber": data.ModelNumber, "Manufacturer": data.Manufacturer}
df = pd.DataFrame(data = dictionary)
counter = 0
print("<string-array name=\"Brand\">")
for index in range(len(df.EquipmentName.unique())):
    names = df.EquipmentName.unique()
    name = names[index]
    modelNumber = df.ModelNumber[index]
    brand = df.Manufacturer[index]
    try:
        name = float(name)
        if math.isnan(name):
            continue
    except ValueError:
        pass
    try:
        modelNumber = float(modelNumber)
        if math.isnan(modelNumber):
            continue
    except ValueError:
        pass
    try:
        brand = float(brand)
        if math.isnan(brand):
            continue
    except ValueError:
        pass
    
    
    counter += 1
    
    
    print("\t<item>" + str(name) + "</item>")
print("</string-array>")
print(counter)
```
